```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
data=pd.read_csv("Iris.csv")#عشان اشتغل ع نفس الداتا هحذف اخر عمود الى هو الy عشان بشتغل unsupervised
data=data.drop(["label"],axis=1)
#analysing
print(data.head())
print(data.isnull().sum())
#scaling
x= data.iloc[:,:-1]
y= data.iloc[:,-1]
```

```python
c_num=[]
j=[]
for i in range(1,10):
    model=KMeans(n_clusters=i)
    model.fit(data)
    c_num.append(i)#حط فيها قيمه الi مع كل لفه
    j.append(model.inertia_)
```

```python
plt.plot(c_num,j)
plt.show()
```
